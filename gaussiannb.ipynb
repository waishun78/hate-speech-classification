```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("./train_tfidf_features.csv")
df
```

```python
label = df["label"]
features_names = [str(i) for i in range(0,5000)]
features = df.loc[:,features_names]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_rescaled = scaler.fit_transform(features)
```

```python
pca_95 = PCA(n_components=0.95)
pca_95_features = pca_95.fit_transform(features_rescaled)
pca_95_features_df = pd.DataFrame(data=pca_95_features)
```

```python
pca_95_features_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
X, y = pca_95_features_df, label
gnb = GaussianNB()
y_pred = gnb.fit(X, y).predict(X)
print(accuracy_score(y, y_pred))
X_rescaled = scaler.fit_transform(pca_95_features_df)
```

```python
df_test = pd.read_csv("./test_tfidf_features.csv")
df_testfeatures = df_test.loc[:, features_names]
testfeatures_rescaled = scaler.fit_transform(df_testfeatures)
pca_95_testfeatures = pca_95.transform(testfeatures_rescaled)
pca_95_testfeatures_df = pd.DataFrame(pca_95_testfeatures)
pca_95_testfeatures_df
```

```python
from sklearn import tree
mnb = tree.DecisionTreeClassifier()
y_test_pred = mnb.fit(X, label).predict(pca_95_testfeatures)
y_test_pred = pd.DataFrame(y_test_pred)

y_test_pred.columns = ["label"]
y_test_pred.insert(0,"id", df_test["id"])
y_test_pred.to_csv('gaussian_nb_predictions.csv', index=False)

y_pred = mnb.predict(X)
print(accuracy_score(y,y_pred))
```
